```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
file_path = '/Users/diya/Python Projects/NSE/NSE 2016-2017.csv'
data = pd.read_csv(file_path)
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Convert TIMESTAMP to datetime
data['TIMESTAMP'] = pd.to_datetime(data['TIMESTAMP'])

# Ensure numeric columns are of the correct type
numeric_columns = ['OPEN', 'HIGH', 'LOW', 'CLOSE', 'LAST', 'PREVCLOSE', 'TOTTRDQTY', 'TOTTRDVAL', 'TOTALTRADES']
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
data[numeric_columns]
```

```python
# Add a column for daily price range
data['PRICE_RANGE'] = data['HIGH'] - data['LOW']
data['PRICE_RANGE']
```

```python
# Add a column for percentage change from previous close to current close
data['PCT_CHANGE'] = ((data['CLOSE'] - data['PREVCLOSE']) / data['PREVCLOSE']) * 100
data['PCT_CHANGE']
```

```python
# Check for duplicates
duplicates = data.duplicated().sum()
duplicates
```

```python
# Remove duplicates
data = data.drop_duplicates()
data
```

```python
# Plot closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(data['TIMESTAMP'], data['CLOSE'], color='skyblue', label='Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Closing Price Over Time')
plt.legend()
plt.show()
```

```python
# Plot the distribution of percentage changes
plt.figure(figsize=(10, 6))
sns.histplot(data['PCT_CHANGE'], bins=50, kde=True)
plt.xlabel('Percentage Change')
plt.title('Distribution of Percentage Changes')
plt.show()
```

```python
# Scatter plot of traded quantity vs. traded value
plt.figure(figsize=(10, 6))
plt.scatter(data['TOTTRDQTY'], data['TOTTRDVAL'], color='purple', alpha=0.5)
plt.xlabel('Total Traded Quantity')
plt.ylabel('Total Traded Value')
plt.title('Traded Quantity vs. Traded Value')
plt.show()
```

```python
# Group by date and calculate the total traded quantity
daily_traded_qty = data.groupby('TIMESTAMP')['TOTTRDQTY'].sum().reset_index()

# Plot the trend of daily traded quantity over time
plt.figure(figsize=(10, 6))
plt.plot(daily_traded_qty['TIMESTAMP'], daily_traded_qty['TOTTRDQTY'], color='seagreen', label='Daily Traded Quantity')
plt.xlabel('Date')
plt.ylabel('Total Traded Quantity')
plt.title('Trend of Daily Traded Quantity Over Time')
plt.legend()
plt.show()
```

```python
# Calculate daily volatility as the percentage range (HIGH - LOW) / LOW
data['VOLATILITY'] = (data['HIGH'] - data['LOW']) / data['LOW'] * 100

# Plot the distribution of volatility
plt.figure(figsize=(10, 6))
sns.histplot(data['VOLATILITY'], bins=50, kde=True, color='orange')
plt.xlabel('Volatility (%)')
plt.title('Distribution of Daily Volatility')
plt.show()
```

```python
# Group by symbol and calculate the mean percentage change
mean_pct_change = data.groupby('SYMBOL')['PCT_CHANGE'].mean().reset_index()

# Sort by percentage change and get the top 10 performers
top_performers = mean_pct_change.sort_values(by='PCT_CHANGE', ascending=False).head(10)

# Plot the top 10 performers
plt.figure(figsize=(12, 6))
sns.barplot(x='PCT_CHANGE', y='SYMBOL', data=top_performers, palette='viridis')
plt.xlabel('Mean Percentage Change')
plt.ylabel('Stock Symbol')
plt.title('Top 10 Performing Stocks Based on Mean Percentage Change')
plt.show()
```

```python
# Extract month and year from the TIMESTAMP column
data['YEAR'] = data['TIMESTAMP'].dt.year
data['MONTH'] = data['TIMESTAMP'].dt.month

# Calculate the monthly average closing price
monthly_avg_close = data.groupby(['YEAR', 'MONTH', 'SYMBOL'])['CLOSE'].mean().reset_index()

# Plot the monthly average closing price for a few selected stocks
selected_stocks = ['20MICRONS', '3IINFOTECH', '3MINDIA']
plt.figure(figsize=(12, 6))
for stock in selected_stocks:
    stock_data = monthly_avg_close[monthly_avg_close['SYMBOL'] == stock]
    plt.plot(stock_data['MONTH'], stock_data['CLOSE'], label=stock)

plt.xlabel('Month')
plt.ylabel('Average Closing Price')
plt.title('Monthly Average Closing Price for Selected Stocks')
plt.legend()
plt.show()
```

```python
# Add hypothetical sector information
import random

sectors = ['Technology', 'Finance', 'Healthcare', 'Energy']
data['SECTOR'] = [random.choice(sectors) for _ in range(len(data))]

# Calculate the average percentage change by sector
sector_performance = data.groupby('SECTOR')['PCT_CHANGE'].mean().reset_index()

# Plot the average percentage change by sector
plt.figure(figsize=(10, 6))
sns.barplot(x='PCT_CHANGE', y='SECTOR', data=sector_performance, palette='pastel')
plt.xlabel('Average Percentage Change')
plt.ylabel('Sector')
plt.title('Average Percentage Change by Sector')
plt.show()
```

```python
# Calculate the percentage change over the last month
recent_data = data[data['TIMESTAMP'] >= (data['TIMESTAMP'].max() - pd.DateOffset(months=1))]
momentum = recent_data.groupby('SYMBOL')['PCT_CHANGE'].sum().reset_index()

# Sort by momentum and get the top 10 stocks
top_momentum_stocks = momentum.sort_values(by='PCT_CHANGE', ascending=False).head(10)

# Plot the top 10 momentum stocks
plt.figure(figsize=(12, 6))
sns.barplot(x='PCT_CHANGE', y='SYMBOL', data=top_momentum_stocks, palette='muted')
plt.xlabel('Total Percentage Change')
plt.ylabel('Stock Symbol')
plt.title('Top 10 Stocks with Highest Momentum Over the Last Month')
plt.show()
```
